# Computational evaluation for neural networks

## Define a traditional 2D convolution calculation function

```python
def Conv_calculation (W , H, Stride, Pad, InChannel, OutChannel, KernelSize, MultiNumber):
    MultiUseNumber = int(MultiNumber/(KernelSize*KernelSize))*KernelSize*KernelSize
    calculation = (W + Pad*2 - (KernelSize - 1)) * (H + Pad*2 - (KernelSize - 1)) \
                  * InChannel * OutChannel * KernelSize * KernelSize \
                  /(Stride * Stride * MultiUseNumber)
    return calculation
```

## Define a Winograd agorithm of 2D convolution calculation function

```python
def Winograd_calculation (W , H, Stride, Pad, InChannel, OutChannel, KernelSize, MultiNumber, nSize, mSize):
    MultiUseNumber = int(MultiNumber/(nSize*nSize))*nSize*nSize
    calculation = (W + Pad*2 - (KernelSize - 1)) * (H + Pad*2 - (KernelSize - 1)) \
                  * InChannel * OutChannel * nSize * nSize \
                  /(Stride * Stride * MultiUseNumber)/(mSize*mSize)
    return calculation
```

## Define an evaluation of 2D convolution occupying memory funtion

```python
def ConvMemory_calculation (W, H, Stride, Pad, InChannel, OutChannel,KernelSize, MutiNumber, nSize):
    ConvImMemory = (W + 2*Pad) * KernelSize * KernelSize * InChannel * 4
    ConvFiMemory = KernelSize * KernelSize * OutChannel * 4
    WinogradImMemory = (W + 2*Pad) * nSize * nSize * InChannel * 4
    WinogradFiMemory = nSize * nSize * OutChannel* 4
    return (ConvImMemory, ConvFiMemory, WinogradImMemory, WinogradFiMemory)
```

## Define Batch and Pooling calculation function

```python
def BatchPooling_calculation (W, H, InChannel):
    BatchPooling_cycle = W * H * InChannel / 4
    DDR_Cycle = BatchPooling_cycle * 8 / 57
    return (BatchPooling_cycle, DDR_Cycle)
```

## Multiplication cycle and read data bandwith

```python
def CycleBandwith (W, Pad, InChannel, OutChannel, KernelSize, MultiNumber, nSize):
    WinoMultiUseNumber = int(MultiNumber/(nSize*nSize))*nSize*nSize
    ConvMultiUseNumber = int(MultiNumber/(KernelSize*KernelSize))*KernelSize*KernelSize
    ConvCycle = OutChannel * KernelSize * KernelSize / ConvMultiUseNumber
    Winocycle = OutChannel * nSize * nSize / WinoMultiUseNumber
    ConvBRamCycle = KernelSize * KernelSize / 4
    WinoBRamCycle = nSize * nSize / 4
    ConvDDRCycle = (KernelSize * KernelSize * 2) * 2 / 57
    WinoDDRCycle = (nSize * nSize * 2) * 2 / 57
    return (ConvCycle, ConvBRamCycle, ConvDDRCycle, Winocycle, WinoBRamCycle, WinoDDRCycle)
```
